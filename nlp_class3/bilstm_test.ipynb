```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, GRU, Bidirectional
```

```python
T = 8
D = 2
M = 3

X = np.random.randn(1, T, D)
```

```python
input_ = Input(shape=(T, D))
# rnn = Bidirectional(LSTM(M, return_state=True, return_sequences=True))
rnn = Bidirectional(LSTM(M, return_state=True, return_sequences=False))
x = rnn(input_)

model = Model(inputs=input_, outputs=x)
o, h1, c1, h2, c2 = model.predict(X)
```

```python
print("o.shape:", o.shape) # h1 and h1 concat 起來了，所以最後一維度是6
print("h1.shape:", h1.shape)
print("c1.shape:", c1.shape)
```

```python
print("o:", o) # h1 and h1 value 接起來成為output，注意在 return_sequences=True，h1會出現在 output 的最後一個step, 而h2(反向)會出現在output 的第一個step
print("h1:", h1)
print("c1:", c1)
print("h2:", h2)
print("c2:", c2)
```
